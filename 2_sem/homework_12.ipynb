```python
def experiment_(num_experiments):
    results = []
    mean_values = []

    for _ in range(num_experiments):
        d4 = random.randint(1, 4)
        d6 = random.randint(1, 6)
        d12 = random.randint(1, 12)
        
        result = d4 + d6 + d12
        results.append(result / 3)
        mean_values.append(sum(results) / len(results))

    return mean_values

y = experiment_(1000)
x = [i for i in range(1, 1000 + 1)]
```

3. –ü–æ–¥–±—Ä–∞—Å—ã–≤–∞—é—Ç—Å—è —á–µ—Ç—ã—Ä–µ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω—ã–µ –∏–≥—Ä–∞–ª—å–Ω—ã–µ –∫–æ—Å—Ç–∏ (—Å —à–µ—Å—Ç—å—é –≥—Ä–∞–Ω—è–º–∏). 
–û–±–æ–∑–Ω–∞—á–∏–º —á–µ—Ä–µ–∑ ùëÜ —Å—É–º–º—É –Ω–∞–∏–±–æ–ª—å—à–∏—Ö —Ç—Ä–µ—Ö –≤—ã–ø–∞–≤—à–∏—Ö –æ—á–∫–æ–≤ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ 
–≤—ã–ø–∞–ª–æ –Ω–∞ –ø–µ—Ä–≤–æ–π –∫–æ—Å—Ç–∏ 1, –Ω–∞ –≤—Ç–æ—Ä–æ–π ‚Äì 2, –Ω–∞ —Ç—Ä–µ—Ç—å–µ–π ‚Äì 6, –Ω–∞ —á–µ—Ç–≤–µ—Ä—Ç–æ–π ‚Äì 1, —Ç–æ–≥–¥–∞ 
–∑–Ω–∞—á–µ–Ω–∏–µ —Ç–∞–∫–æ–π —Å—É–º–º—ã —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 9). –ù–∞–π–¥–∏—Ç–µ (–∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏) –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ 
–æ–∂–∏–¥–∞–Ω–∏–µ ùê∏(ùëÜ). –†–∞–∑—ã–≥—Ä–∞–π—Ç–µ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç –Ω–∞ Python –∏ –Ω–∞–π–¥–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ 
—Ç–∞–∫–æ–π —Å—É–º–º—ã. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è —Å—É–º–º—ã –æ—Ç —á–∏—Å–ª–∞ 
—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤.

1) –°—É–º–º—ã –±—É–¥—É—Ç –ø–æ–ø–∞–¥–∞—Ç—å —á—ë—Ç–∫–æ –≤ –¥–∏–∞–ø–∞–∑–æ–Ω –æ—Ç 3-—Ö –¥–æ 18-—Ç–∏ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ
2) –ù–∞–π–¥—ë–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã–ø–∞–¥–µ–Ω–∏—è –∫–∞–∂–¥–æ–π —Å—É–º–º—ã, –∑–∞—Ç–µ–º –ø–æ—Å—á–∏—Ç–∞–µ–º E(S) –ø–æ –∏–∑–≤–µ—Å—Ç–Ω–æ–π —Ñ–æ—Ä–º—É–ª–µ –º–∞—Ç. –æ–∂–∏–¥–∞–Ω–∏—è:

E(S) = 3 * P(3) + 4 * P(4) + ... + 18 * P(18) = 12.2446

```python
def experiment_(num_experiments):
    results = []
    mean_values = []

    for _ in range(num_experiments):
        d1 = random.randint(1, 6)
        d2 = random.randint(1, 6)
        d3 = random.randint(1, 6)
        d4 = random.randint(1, 6)

        list_ = sorted([d1, d2, d3, d4])[::-1]
        
        result = sum(list_[:3])
        results.append(result / 3)
        mean_values.append(sum(results) / len(results))

    return mean_values

y = experiment_(1000)
x = [i for i in range(1, 1000 + 1)]
```

```python
def plot_(x, y):
    plt.plot(x, y, color = 'b')
    plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Å–ø—ã—Ç–∞–Ω–∏–π')
    plt.ylabel('–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—É–º–º—ã')
    plt.grid()
    plt.show()


plot_(x, y)
```

4. –ü–æ–¥–±—Ä–∞—Å—ã–≤–∞—é—Ç—Å—è —á–µ—Ç—ã—Ä–µ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω—ã–µ –∏–≥—Ä–∞–ª—å–Ω—ã–µ –∫–æ—Å—Ç–∏ (—Å —à–µ—Å—Ç—å—é –≥—Ä–∞–Ω—è–º–∏). 
–û–±–æ–∑–Ω–∞—á–∏–º —á–µ—Ä–µ–∑ ùëã –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤—ã–ø–∞–≤—à–∏—Ö –æ—á–∫–æ–≤ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ 
–≤—ã–ø–∞–ª–æ –Ω–∞ –ø–µ—Ä–≤–æ–π –∫–æ—Å—Ç–∏ 1, –Ω–∞ –≤—Ç–æ—Ä–æ–π ‚Äì 2, –Ω–∞ —Ç—Ä–µ—Ç—å–µ–π ‚Äì 6, –Ω–∞ 1 —á–µ—Ç–≤–µ—Ä—Ç–æ–π ‚Äì 1, —Ç–æ–≥–¥–∞ 
–º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 6). –ù–∞–π–¥–∏—Ç–µ (–∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏) –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ 
–æ–∂–∏–¥–∞–Ω–∏–µ ùê∏(ùëã) –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—é ùëâùëéùëü(ùëã). –†–∞–∑—ã–≥—Ä–∞–π—Ç–µ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç –Ω–∞ Python –∏ –Ω–∞–π–¥–∏—Ç–µ 
—Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ç–∞–∫–æ–π –≤–µ–ª–∏—á–∏–Ω—ã. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —Å—Ä–µ–¥–Ω–µ–≥–æ 
–∑–Ω–∞—á–µ–Ω–∏—è –æ—Ç —á–∏—Å–ª–∞ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤.

1) –ó–Ω–∞—á–µ–Ω–∏—è –±—É–¥—É—Ç —á—ë—Ç–∫–æ –ø–æ–ø–∞–¥–∞—Ç—å –≤ –≥—Ä–∞–Ω–∏—Ü—ã –æ—Ç 1 –¥–æ 6-—Ç–∏ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.
2) –ù–∞–π–¥—ë–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã–ø–∞–¥–µ–Ω–∏—è –∫–∞–∂–¥–æ–π —Å—É–º–º—ã, –∑–∞—Ç–µ–º –ø–æ—Å—á–∏—Ç–∞–µ–º E(X) –∏ Var(X) –ø–æ –∏–∑–≤–µ—Å—Ç–Ω—ã–º —Ñ–æ—Ä–º—É–ª–∞–º –º–∞—Ç. –æ–∂–∏–¥–∞–Ω–∏—è –∏ –¥–∏—Å–ø–µ—Ä—Å–∏–∏:

E(X) = 1 * P(1) + 2 * P(2) + ... + 6 * P(6) = 5.2446

Var(X) = E(X^2) - (E(X))^2 = 0.910079

```python
def experiment_(num_experiments):
    results = []
    mean_values = []

    for _ in range(num_experiments):
        d1 = random.randint(1, 6)
        d2 = random.randint(1, 6)
        d3 = random.randint(1, 6)
        d4 = random.randint(1, 6)

        result = max([d1, d2, d3, d4])
        results.append(result)
        mean_values.append(sum(results) / len(results))

    return mean_values

y = experiment_(1000)
x = [i for i in range(1, 1000 + 1)]
```

```python
def plot_(x, y):
    plt.plot(x, y, color = 'b')
    plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Å–ø—ã—Ç–∞–Ω–∏–π')
    plt.ylabel('–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—É–º–º—ã')
    plt.grid()
    plt.show()


plot_(x, y)
```
