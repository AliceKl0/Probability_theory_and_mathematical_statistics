```python
 def experiment_(num_experiments):
    mean_values = []
    results_list = []
    
    for _ in range(num_experiments):
        
        m1 = random.randint(0, 1)
        m2 = random.randint(0, 1)
        i_ = 0
        results = 0
        
        while not (m1 == 1 and m2 == 1):

            results += 1
            m1 = random.randint(0, 1)
            m2 = random.randint(0, 1)
        
        results_list.append(results)
        mean_values.append(sum(results_list) / len(results_list))

    return mean_values

y = experiment_(1000)
x = [i for i in range(1, 1000 + 1)]
```

```python
def plot_(x, y):
    plt.plot(x, y, color = 'b')
    plt.xlabel('Количество испытаний')
    plt.ylabel('Среднее значение кол-ва бросков')
    plt.grid()
    plt.show()


plot_(x, y)
```

2. Предположим, что игральная кость не является симметричной и массы в ней 
распределены так, что масса каждой грани пропорциональна её номеру. 
Несимметричная игральная кость бросается до тех пор, пока не выпадут все четные 
цифры. Найдите математичеcкое ожидание и дисперсию числа бросков. Постройте 
график зависимости среднего значения числа бросков от числа экспериментов.

```python
xi = [1, 2, 3, 4, 5, 6]
px = [2 * i / (6 * (6 + 1)) for i in range(1, 6 + 1)]

p_chet = px[1] * px[3] * px[5] * 6 # 6 вариантов порядка выпадения 2, 4, и 6 (3!)
p_chet
```

```python
X = sts.geom(p_chet)
```

```python
print(X.mean())
print(X.var())
```
