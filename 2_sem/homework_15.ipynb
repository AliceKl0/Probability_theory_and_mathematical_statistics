```python
import numpy as np
import scipy.stats as sts
import sympy as sp
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
import random
```

```python
xi = [1, 2, 3, 4, 5]
pi = [1/6, 1/6, 1/3, 1/6, 1/6]
```

```python
X = sts.rv_discrete(values=(xi, pi))
```

```python
nu1 = X.moment(1)
nu2 = X.moment(2)
nu3 = X.moment(3)
nu4 = X.moment(4)
nu1, nu2, nu3, nu4
```

```python
yi = [xi - X.mean() for xi in xi]
yi
```

```python
Y = sts.rv_discrete(values=(yi, pi))
```

```python
mu2 = Y.moment(2)
mu3 = nu3 - 3 * nu1 * nu2 + 2 * nu1**3
mu4 = nu4 - 4 * nu1 * nu3 + 6 * nu1**2 * nu2 - 3 * nu1**4
mu2, mu3, mu4
```

```python
asx =  (1 / 24**0.5) * (mu3 / (mu2**0.5)** 3)
asx
```

```python
ex = (1/24) * (mu4 / (mu2**0.5)**4 - 3)
ex
```
