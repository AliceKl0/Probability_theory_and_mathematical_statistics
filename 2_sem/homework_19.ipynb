```python
import scipy.integrate as integrate
import sympy as sp
import math
```

```python
def target_function_f1(x):
 return (1 + 3*x**0.5 + 4*x**0.7 + 9*x**0.9)**1.1
```

```python
E_x = integrate.quad(target_function_f1, 4, 7)[0] * 1/3
E_x # Мат. ожидание
```

```python
def target_function_f2(x):
 return ((1 + 3*x**0.5 + 4*x**0.7 + 9*x**0.9)**1.1)**2
```

```python
E_x2 = integrate.quad(target_function_f2, 4, 7)[0] * 1/3
E_x2
```

```python
sqrt_Var = math.sqrt(E_x2 - E_x)
sqrt_Var # Стандартное отклонение
```

```python
def target_function_f3(x):
 return ((1 + 3*x**0.5 + 4*x**0.7 + 9*x**0.9)**1.1)**3
```

```python
E_x3 = integrate.quad(target_function_f3, 4, 7)[0] * 1/3
E_x3
```

```python
u3 = E_x3 - 3*E_x*E_x2 + 2*E_x**3
```

```python
Asm = u3 / sqrt_Var**3
Asm  # Асимметрия
```
