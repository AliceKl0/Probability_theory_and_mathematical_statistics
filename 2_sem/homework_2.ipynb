5.
а) Найти для n = 3 обратную матрицу Гильберта.
б) Найти матрицу Гильберта для n = 5.

```python
# а)

import math

def hilb(n):
    h = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            h[i][j] = 1 / (i + j + 1)
    return h

h = hilb(3)

def obr_hilb(h):
    h_1 = h.copy()
    n = len(h)
    
    for i in range(n):
        for j in range(n):
            h_1[i][j] = (-1)**(i + j + 2) * (i + j + 1) * math.comb(n + i, n - (j + 1)) * math.comb(n + j, n - (i + 1)) * (math.comb(i + j, i))**2
    return h_1

h_1 = obr_hilb(h)

for row in h_1:
    print(f"({' '.join([str(el) for el in row])})")
```

```python
# б)

def hilb(n):
    h = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            h[i][j] = f"1/{i + j + 1}" if not (i == 0 and j == 0) else '1'
    return h

h = hilb(5)

for row in h:
    print(f"({' '.join(row)})")
```

6. Используя графические возможности, изобразите матрицу размера n × n для n = 100,
каждая строка которой содержит элементы биномиальных коэффициентовпо модулю 2
(закрашенный «синий» квадратик соответствует 1, «желтый»
закрашенный – 0).

```python
import matplotlib.pyplot as plt
import numpy as np

n = 100
mtr = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        mtr[i][j] = np.math.comb(i, j) % 2

plt.imshow(mtr, cmap='viridis')

plt.show()
```
