1. Иван и Пётр договорились о встрече в определённом месте между 11 и 12 часами. 
Каждый приходит и ждёт другого до истечения часа, но не более 15 минут. Найдите 
вероятности следующих событий:

```python
# A – встреча состоялась

import math
import numpy as np
import random

def stat_prob(N):
    M = 0
    a = 15
    x = np.random.uniform(0, 60, N)
    y = np.random.uniform(0, 60, N)

    for i in range(N):
        if (0 <= y[i] - x[i] <= a) or (0 <= x[i] - y[i] <= a):
            M += 1
    
    return M/N

pA = (60 ** 2 - 45 ** 2) / 60 ** 2
print(pA)
```

```python
import matplotlib.pyplot as plt

def plot_(n):
    x = [i for i in range(1, n + 1)]
    y = [stat_prob(num) for num in x]
    plt.plot(x, y, color = 'b')
    plt.xlabel('Количество испытаний')
    plt.ylabel('Вероятность события')
    plt.grid()
    plt.plot(x, [pA] * len(x), color = 'r')
    plt.title('A – встреча состоялась')
    plt.show()


plot_(1000)
```

3. (С использованием «Jupyter Notebook» или «Wolfram Mathematica») 
Рассматривается отрезок [0; 1] и принадлежащий ему отрезок [α, β] = [0, 5; 0, 75]. 
Эксперимент состоит в том, что в системе «Jupytet Notebook» или «Wolfram 
Mathematica» выполняется команда np.random.uniform(0,1) или RandomReal[ ]. 
Интересующее нас событие (A) – результат выполнения команды попадает на 
отрезок [α, β]. Эксперимент проводится 𝑁 = 1000; 10 000; 100 000; 1 000 000;
10 000 000 раз. Требуется определить соответствующие относительные частоты 𝑁𝐴
𝑁
рассматриваемого события. Построить график зависимости относительной частоты 
от 𝑁.

```python
import math
import numpy as np
import random

alpha = 0.5
beta = 0.75

N_values = [1000, 10000, 100000, 1000000, 10000000]
values = []

for N in N_values:
    count = 0
    
    for _ in range(N):
        result = np.random.uniform(0, 1)
        
        if alpha <= result <= beta:
            count += 1
    
    values.append(count / N)

for i in range(len(values)):
    print(f"Для {N_values[i]}: p = {values[i]}")
```

```python
import matplotlib.pyplot as plt

plt.plot(N_values, values, marker='o')
plt.xscale('log')
plt.xlabel('Число экспериментов')
plt.ylabel('Относительная частота')
plt.title('Зависимость относительной частоты от числа экспериментов')
plt.show()
```

4. (С использованием «Jupyter Notebook» или «Wolfram Mathematica») Отрезок [0; 1] 
разбивается на 15 одинаковых отрезков. Наудачу выбирается из них 3 отрезка. 
Вычисляется значение функции np.random.uniform(0,1) или RandomReal[ ]. Найти 
вероятность того, что это значение будет принадлежать одному из выбранных 
отрезков. Эксперимент проводится 𝑁 = 1000, 10 000, 100 000, 1 000 000,
10 000 000 раз. Найти соответствующую статистическую вероятность и построить 
график зависимости относительной частоты от числа экспериментов 𝑁. 

```python
import numpy as np

N_values = [1000, 10000, 100000, 1000000, 10000000]
num_segments = 15
num_selected_segments = 3
values = []
segment_boundaries = [[i/num_segments, (i+1)/num_segments] for i in range(num_segments)] # Определяем границы сегментов

count = 0

for num in range(10000001):
    selected_segments_num = np.random.choice(range(num_segments), num_selected_segments, replace=False)  # Выбираем 3-ри номера сегментов
    selected_segments = [segment_boundaries[ind] for ind in selected_segments_num]
    rand_value = np.random.uniform(0, 1)
        
    for segment in selected_segments:
        if segment[0] <= rand_value < segment[1]:
            count += 1
            break
    
    if num in N_values:
        values.append(count / num)

for i in range(len(values)):
    print(f"Для {N_values[i]}: p = {values[i]}")
```

```python
import matplotlib.pyplot as plt

plt.plot(N_values, values, marker='o')
plt.xscale('log')
plt.xlabel('Число экспериментов')
plt.ylabel('Относительная частота')
plt.title('Зависимость относительной частоты от числа экспериментов')
plt.show()
```

5. Отрезок [0; 1] случайным образом делится на 3 части. Найти вероятность того, 
что из полученных отрезков можно сложить треугольник и этот треугольник будет 
остроугольным. Найти соответствующую статистическую вероятность и показать 
статистическую устойчивость вероятности в зависимости от числа экспериментов 
𝑁.

```python
import math


def stat_prob(N):
    M = 0

    for i in range(N):
        x, y = sorted(np.random.uniform(0, 1, 2))

        if (x <= y) and (y >= 1/2) and (x <= 1/2) and (y <= x + 1/2) and (x ** 2 + (1 - y) ** 2 >= (y - x) ** 2) and (x ** 2 + (y - x) ** 2 >= (1 - y) ** 2) and ((y - x) ** 2 + (1 - y) ** 2 >= x ** 2):
            M += 1
    
    return M/N

p = (3/2 * math.log(2) - 1) / (1 / 2)
print(p)
```

```python
def plot_(n):
    x = [i for i in range(1, n + 1)]
    y = [stat_prob(num) for num in x]
    plt.plot(x, y, color = 'b')
    plt.xlabel('Количество испытаний')
    plt.ylabel('Вероятность события')
    plt.grid()
    plt.plot(x, [p] * len(x), color = 'r')
    plt.title('Статистическая устойчивость вероятности')
    plt.show()


plot_(1000)
```
