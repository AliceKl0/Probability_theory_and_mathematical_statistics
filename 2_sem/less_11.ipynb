```python
import numpy as np
import scipy.stats as sts
import sympy as sp
import matplotlib.pyplot as plt
import math
```

```python
xi = [1, 2, 3]
Px = [1/3] * 3
```

```python
X = sts.rv_discrete(name='X', values=(xi, Px))
```

```python
X.pmf(xi[0])
```

```python
X.cdf(xi[1])
```

```python
X.cdf(2) - X.cdf(1) 
```

```python
plt.plot(xi, X.pmf(xi), 'o', ms=12, color='r')
plt.vlines(xi, 0, X.pmf(xi), linestyles='dashed', color='r')
plt.hlines(0, min(xi), max(xi), color='b')
```

```python
D = sp.Matrix([xi, Px])
D
```

```python
F = np.insert(Px, 0, 0)

for i in range (1, len(Px) + 1):
    F[i] = F[i - 1] + F[i]

F
```

```python
plt.step(np.append(xi, 8), F, color='r')
```
