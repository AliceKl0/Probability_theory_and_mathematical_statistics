```python
import numpy as np
import scipy.stats as sts
import scipy.optimize as opt
import scipy.integrate as quad
import sympy as sp
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
import random
```

# а)

```python
a = 2
b = 3
X = sts.uniform(a, b - a)
```

```python
X.mean()
```

```python
X.var()
```

```python
x = np.linspace(0, 5, 100)
y = X.pdf(x)
plt.grid()
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.title('График плотности равномерного распределения на [2; 3]')
plt.plot(x, y)
plt.show()
```

```python
# Ф-ция распредлеения

x = np.linspace(0, 5, 100)
y = X.cdf(x)
plt.grid()
plt.xlabel('x')
plt.ylabel('y = F(x)')
plt.title('График ф-ции распределения на [2; 3]')
plt.plot(x, y)
plt.show()
```

# в)

```python
a = 0
sigma = 1
Z = sts.norm(a, sigma)
```

```python
x = np.linspace(-5, 5, 100)
y = Z.pdf(x)
plt.grid()
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.title('График плотности нормального распределения')
plt.plot(x, y)
plt.show()
```

```python
# sigma - фикс.

x = np.linspace(-5, 5, 100)
sigma = 1

for a in -2, 0, 2:
    y = sts.norm(a, sigma).pdf(x)
    plt.plot(x, y)

plt.grid()
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.title('График плотности нормального распределения')
plt.plot(x, y)
plt.show()
```

```python
# a - фикс.

x = np.linspace(-5, 5, 100)
a = 0

for sigma in 1, 2, 0:
    y = sts.norm(a, sigma).pdf(x)
    plt.plot(x, y)

plt.grid()
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.title('График плотности нормального распределения')
plt.plot(x, y)
plt.show()
```
