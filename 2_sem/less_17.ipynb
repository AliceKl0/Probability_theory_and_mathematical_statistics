```python
a = 2
b = 3
X = sts.uniform(a, b - a)
```

```python
X.mean()
```

```python
x = np.linspace(0, 5, 100)
y = X.pdf(x)
plt.grid()
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.title('     [2; 3]')
plt.plot(x, y)
plt.show()
```

```python
# - 

x = np.linspace(0, 5, 100)
y = X.cdf(x)
plt.grid()
plt.xlabel('x')
plt.ylabel('y = F(x)')
plt.title(' -   [2; 3]')
plt.plot(x, y)
plt.show()
```

```python
# sigma - .

x = np.linspace(-5, 5, 100)
sigma = 1

for a in -2, 0, 2:
    y = sts.norm(a, sigma).pdf(x)
    plt.plot(x, y)

plt.grid()
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.title('   ')
plt.plot(x, y)
plt.show()
```

```python
# a - .

x = np.linspace(-5, 5, 100)
a = 0

for sigma in 1, 2, 0:
    y = sts.norm(a, sigma).pdf(x)
    plt.plot(x, y)

plt.grid()
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.title('   ')
plt.plot(x, y)
plt.show()
```
